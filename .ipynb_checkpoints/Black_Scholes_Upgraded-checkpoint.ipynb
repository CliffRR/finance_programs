```python
import yahoo_fin.stock_info as si
import yahoo_fin.options as ops
import numpy as np 
import pandas as pd
import math
from datetime import datetime
```

```python
#---------------------------------------------Location for Functions--------------------------------------------------

# Python program to convert a list to string     
# Function to convert  
def listToString(List_Name): 
    
    # initialize an empty string
    string = "" 
    
    # traverse in the string  
    for i in List_Name: 
        string += i  
    
    # return string  
    return string
        
#Program to remove a certain part of a string 
#Still needs to be fixed
def RemoveString(String_Name, Remove_String_Part, Empty_List_Name):
    
    for i in String_Name:
        if i == Remove_String_Part:
            break
        Empty_List_Name.append(i)
```

```python
#---------------------------------------------BlackScholes Functions-------------------------------------------------- 
#Formula for d1

def BlackScholesCall(p,s,v,t,r):
    A = math.log(p/s) #1st part of the numerator
    B = ((r + ((v ** 2)/2)) * t) #2nd part of the numerator  
    C = (v * math.sqrt(t)) #the denominator

    d1 = ((A + B) / C)
    d2 = float(d1 - C) 


#Formula for y1
    E = float(.2316419 * abs(d1))
    y1 = float((1 / (1 + E))) 
 
#Formula for y2
    F = float(.2316419 * abs(d2))
    y2 = float((1 / (1 + F)))


#Formula for z1
    G = (-1 * ((d1 ** 2)/2))
    z1 = float(.3989423 * (math.exp(G)))  

#Formula for z2
    H = (-1 * ((d2 ** 2)/2))
    z2 = float(.3989423 * (math.exp(H))) 


    I = 1.330274 * (y1 ** 5)
    J = 1.821256 * (y1 ** 4)
    K = 1.781478 * (y1 ** 3)
    L = .356538 * (y1 ** 2)
    M = .3193815 * y1 
    x1 = 1 - (z1 * (I - J + K - L + M))


    I1 = 1.330274 * (y2 ** 5)
    J1 = 1.821256 * (y2 ** 4)
    K1 = 1.781478 * (y2 ** 3)
    L1 = .356538 * (y2 ** 2)
    M1 = .3193815 * y2 
    x2 = 1 - (z2 * (I1 - J1 + K1 - L1 + M1))


    d3 = 1 - x1
    d4 = 1 - x2


    N = (-1 * r * t)
    O = p * x1
    O1 = p * d3
    P = s * math.exp(N) * x2
    PP = s * math.exp(N)
    P1 = s * math.exp(N) * d4
    

    if d1 < 0 and d2 < 0:
        Vc = O1 - P1
        Vp = Vc + PP - p
        
        return Vc
        
        
    if d1 > 0 and d2 > 0:
        Vc = O - P
        Vp = Vc + PP - p
        
        return Vc
        
        
    if d1 > 0 and d2 > 0:
        Vc = O - P
        Vp = Vc + PP - p
        
        return Vc
        
        
    if d1 < 0 and d2 > 0:
        Vc = O1 - P
        Vp = Vc + PP - p     
        
        return Vc
    
    
#Function for Black Scholes Put
def BlackScholesPut(p,s,v,t,r):
    A = math.log(p/s) #1st part of the numerator
    B = ((r + ((v ** 2)/2)) * t) #2nd part of the numerator  
    C = (v * math.sqrt(t)) #the denominator

    d1 = ((A + B) / C)
    d2 = float(d1 - C) 


#Formula for y1
    E = float(.2316419 * abs(d1))
    y1 = float((1 / (1 + E))) 
 
#Formula for y2
    F = float(.2316419 * abs(d2))
    y2 = float((1 / (1 + F)))


#Formula for z1
    G = (-1 * ((d1 ** 2)/2))
    z1 = float(.3989423 * (math.exp(G)))  

#Formula for z2
    H = (-1 * ((d2 ** 2)/2))
    z2 = float(.3989423 * (math.exp(H))) 


    I = 1.330274 * (y1 ** 5)
    J = 1.821256 * (y1 ** 4)
    K = 1.781478 * (y1 ** 3)
    L = .356538 * (y1 ** 2)
    M = .3193815 * y1 
    x1 = 1 - (z1 * (I - J + K - L + M))


    I1 = 1.330274 * (y2 ** 5)
    J1 = 1.821256 * (y2 ** 4)
    K1 = 1.781478 * (y2 ** 3)
    L1 = .356538 * (y2 ** 2)
    M1 = .3193815 * y2 
    x2 = 1 - (z2 * (I1 - J1 + K1 - L1 + M1))


    d3 = 1 - x1
    d4 = 1 - x2


    N = (-1 * r * t)
    O = p * x1
    O1 = p * d3
    P = s * math.exp(N) * x2
    PP = s * math.exp(N)
    P1 = s * math.exp(N) * d4
    
    if d1 < 0 and d2 < 0:
        Vc = O1 - P1
        Vp = Vc + PP - p
        
        return Vp
        
        
    if d1 > 0 and d2 > 0:
        Vc = O - P
        Vp = Vc + PP - p
        
        return Vp
        
        
    if d1 > 0 and d2 > 0:
        Vc = O - P
        Vp = Vc + PP - p
        
        return Vp
        
        
    if d1 < 0 and d2 > 0:
        Vc = O1 - P
        Vp = Vc + PP - p     
        
        return Vp
```

```python
#Variables
#Look at the T-Bill that has the same expiration and days remaining as the contract 

risk_free_interest_rate = .015
ticker_symbol = 'MU'
expiration_date = '6/04/2021'
year = 2021
month = 6
day = 4
Implied_Volatility_List = []

#get the live price 
ticker_symbol_live_price = si.get_live_price(ticker_symbol)
ticker_symbol_live_price
```

```python
#so i can see what row my strike price is at

stock_call_options = ops.get_calls(ticker_symbol, expiration_date)
stock_call_options
```

```python
stock_put_options = ops.get_puts(ticker_symbol, expiration_date)
stock_put_options
```

```python
#Figure out which strike im working with and then use the row associated with it
#Future references: Volume = 8

Row_of_Strike_Put = 1
strike_price_put = stock_put_options.iloc[Row_of_Strike_Put][2]
Implied_Volatility_Put = stock_put_options.iloc[Row_of_Strike_Put][10]

Row_of_Strike_Call = 13
strike_price_call = stock_call_options.iloc[Row_of_Strike_Call][2]
Implied_Volatility_Call = stock_call_options.iloc[Row_of_Strike_Call][10]
#Implied_Volatility_Call
```

```python
#Im getting rid of the percent (%) and turning it into a decimal 

Actual_Implied_Volatility_Call = Implied_Volatility_Call.replace('%','')
Actual_Implied_Volatility_Call = float(Actual_Implied_Volatility_Call)
Actual_Implied_Volatility_Call = Actual_Implied_Volatility_Call/100
#Actual_Implied_Volatility_Call

Actual_Implied_Volatility_Put = Implied_Volatility_Put.replace('%','')
Actual_Implied_Volatility_Put = float(Actual_Implied_Volatility_Put)
Actual_Implied_Volatility_Put = Actual_Implied_Volatility_Put/100
#Actual_Implied_Volatility_Put
```

```python
Days_Remaining_Expiration_List = []
expiration = datetime(year, month, day)
difference = expiration - datetime.now() 
left_over_days = str(difference)
# Ex: 8954 days, 7:03:45.765329
```

```python
#To remove the space in left_over_days

for i in left_over_days:
    if i == ' ':
        break
    Days_Remaining_Expiration_List.append(i)
    
Days_Remaining_Expiration =int(listToString(Days_Remaining_Expiration_List))

Days_Remaining_Fraction = Days_Remaining_Expiration/365
```

# The Black Scholes model tends to over price ITM calls and under price OTM calls
#Actual_Implied_Volatility_Call = .3926

```python
#making the function equal to a certain variable
Value_of_Put = BlackScholesPut(ticker_symbol_live_price, strike_price_put, Actual_Implied_Volatility_Put, Days_Remaining_Fraction, risk_free_interest_rate)
Value_of_Call = BlackScholesCall(ticker_symbol_live_price, strike_price_call, Actual_Implied_Volatility_Call, Days_Remaining_Fraction, risk_free_interest_rate)


#Printing out the value of the Calls and/or Puts
Answer1 = str(input('Would you like to figure out the value of a Call, Put, or Both? '))
if Answer1 == 'Put':
    print('Option Price (Put) = '+ str(Value_of_Put))
if Answer1 == 'Call':
    print('Option Price (Call) = '+ str(Value_of_Call))
elif Answer1 == 'Both':
    print('Option Price (Call) = '+ str(Value_of_Call))
    print('Option Price (Put) = '+ str(Value_of_Put))

```
