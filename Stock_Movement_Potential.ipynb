```python
from collections import defaultdict
import yahoo_fin.stock_info as si
import yahoo_fin.options as ops
import numpy as np 
import pandas as pd
import math
from datetime import datetime
```

```python
#---------------------------------------------Location for Functions--------------------------------------------------

# Python program to convert a list to string     
# Function to convert  
def listToString(List_Name): 
    
    # initialize an empty string
    string = "" 
    
    # traverse in the string  
    for i in List_Name: 
        string += i  
    
    # return string  
    return string
```

```python
def StockPotentialUpwards(Current_stock_price, standard_deviations, volatility, time):
    Volatility_with_Time = (volatility * math.sqrt(time)) 
    
    Expected_Stock_Price = Current_stock_price * math.exp(standard_deviations * Volatility_with_Time)
    return Expected_Stock_Price


def StockPotentialDownwards(Current_stock_price, standard_deviations, volatility, time):
    Volatility_with_Time = (volatility * math.sqrt(time)) 
    
    Expected_Stock_Price = Current_stock_price * math.exp((-standard_deviations) * Volatility_with_Time)
    return Expected_Stock_Price
```

```python
#Variables

ticker_symbol = 'NKE'
live_price = si.get_live_price(ticker_symbol)
standard_deviations_allowed_for_movement = 1

expiration_date = '07/30/2021'
year = int(expiration_date[6:10])
month = int(expiration_date[0:2])
day = int(expiration_date[3:5])
```

```python
stock_call_options = ops.get_calls(ticker_symbol,expiration_date)
stock_put_options = ops.get_puts(ticker_symbol, expiration_date)
#stock_call_options
stock_put_options
```

```python
#Figure out which strike im working with and then use the row associated with it
#Future references: Volume = 8

Row_of_Strike_Put = 32
strike_price_put = stock_put_options.iloc[Row_of_Strike_Put][2]
Implied_Volatility_Put = stock_put_options.iloc[Row_of_Strike_Put][10]
#Implied_Volatility_Put

Row_of_Strike_Call = 5
strike_price_call = stock_call_options.iloc[Row_of_Strike_Call][2]
Implied_Volatility_Call = stock_call_options.iloc[Row_of_Strike_Call][10]
#Implied_Volatility_Call
```

```python
#Im getting rid of the percent (%) and turning it into a decimal 

Actual_Implied_Volatility_Call = Implied_Volatility_Call.replace('%','')
Actual_Implied_Volatility_Call = float(Actual_Implied_Volatility_Call)
Actual_Implied_Volatility_Call = Actual_Implied_Volatility_Call/100
#Actual_Implied_Volatility_Call 

Actual_Implied_Volatility_Put = Implied_Volatility_Put.replace('%','')
Actual_Implied_Volatility_Put = float(Actual_Implied_Volatility_Put)
Actual_Implied_Volatility_Put = Actual_Implied_Volatility_Put/100
#Actual_Implied_Volatility_Put
```

```python
Days_Remaining_Expiration_List = []
expiration = datetime(year, month, day)
difference = expiration - datetime.now() 
left_over_days = str(difference)
# Ex: 8954 days, 7:03:45.765329
```

```python
#To remove the space in left_over_days

for i in left_over_days:
    if i == ' ':
        break
    Days_Remaining_Expiration_List.append(i)
    
Days_Remaining_Expiration =int(listToString(Days_Remaining_Expiration_List))

Days_Remaining_Fraction = Days_Remaining_Expiration/365
```

```python
Answer = str(input("Upward or Downward stock movement? "))
if Answer == 'Upward':
    print('The stock would move up to approximately ' + str(StockPotentialUpwards(live_price, standard_deviations_allowed_for_movement, Actual_Implied_Volatility_Call, Days_Remaining_Fraction)) + ' if it moved ' + str(standard_deviations_allowed_for_movement) + ' standard deviations in exactly ' + str(Days_Remaining_Expiration) + ' days.')
    
elif Answer == 'Downward':
    print('The stock would move down to approximately ' + str(StockPotentialDownwards(live_price, standard_deviations_allowed_for_movement, Actual_Implied_Volatility_Put, Days_Remaining_Fraction)) + ' if it moved ' + str(standard_deviations_allowed_for_movement) + ' standard deviations in exactly ' + str(Days_Remaining_Expiration) + ' days.')
```
